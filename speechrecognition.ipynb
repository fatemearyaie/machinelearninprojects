```python
r = sr.Recognizer()
```

# convert speech to text in real time

```python
while True:
    with sr.Microphone() as source:
        #clear background noise:
        r.adjust_for_ambient_noise(source, duration=0.3)
        
        print("speak now")
        
        #capture the audio:
        audio = r.listen(source)
        
        try:
            text = r.recognize_google(audio)
            print("speaker", text)
        except:
            print("say it again")
```
